```python
import pandas as pd
import scipy.spatial.distance as sd
```

```python
path = 'exp2GO/data/'
```

```python
eisen = pd.read_csv(path+'pnas_95_25_14863__2.csv.zip', skiprows=[0,1], index_col=0)
display(eisen.shape)
display(eisen.columns)
display(eisen.head())
```

```python
eisen = eisen.dropna(axis=0)
eisen.shape
```

```python
eisen = eisen.drop(columns='NAME')
eisen
```

```python
#dist_name = 'euclidean'
#dist_name = 'seuclidean'
dist_name = 'cosine'
#dist_name = 'correlation'
d = sd.pdist(eisen, dist_name)
```

```python
ds = sd.squareform(d)
eisen_dist = pd.DataFrame(ds, index=eisen.index, columns=eisen.index)
display(eisen_dist)
```

```python
eisen_dist.to_csv(path+'yeast_expression_dist_'+dist_name+'.csv')
```
